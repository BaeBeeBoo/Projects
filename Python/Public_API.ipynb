```python
import requests

url = "https://api.sampleapis.com/switch/games"
resp = requests.get(url)

#check status code
resp.status_code
```

```python
import time

name = []
genre = []
developers = []
publishers = []

for i in range(20):
    url = f"https://api.sampleapis.com/switch/games={i+1}"
    if resp.status_code == 200:
        json_data = resp.json()
        name.append(json_data[i]["name"]),
        genre.append(json_data[i]["genre"]),
        developers.append(json_data[i]["developers"]),
        publishers.append(json_data[i]["publishers"])
    else:
        name.append(json_data["error"])
        genre.append(json_data["error"])
        developers.append(json_data["error"])
        publishers.append(json_data["error"])

time.sleep(1)
```

```python
import pandas as pd

data = {"Name": name, "Genre": genre, "Developers": developers, "Publishers": publishers}
games = pd.DataFrame.from_dict(data, orient='index')
games = games.transpose()
games
```
